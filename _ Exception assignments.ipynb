Q2. Write a python program to print Python Exception Hierarchy.

```python
import builtins

def print_exception_hierarchy():
    def print_exception_tree(cls, indent=0):
        print(' ' * indent + cls.__name__)
        for subclass in cls.__subclasses__():
            print_exception_tree(subclass, indent + 4)

    print_exception_tree(builtins.BaseException)

print_exception_hierarchy()
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans:- The ArithmeticError class is a base class for various arithmetic errors in Python. It is the parent class of several other error classes, including ZeroDivisionError, OverflowError, and FloatingPointError.
Here are two examples of arithmetic errors.

1 ZeroDivisionError: This error occurs when you try to divide a number by zero.

2 OverflowError: This error occurs when you perform an arithmetic operation that results in a value that is too large to be represented within the available memory space.


```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
try:
    result = 2 ** 1000
except OverflowError as e:
    print(f"Error: {e}")
```

The code above attempts to calculate 2 raised to the power of 1000. However, the resulting value is too large to fit into the memory allocated for storing numbers.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans:- The LookupError class is used in Python to handle exceptions that occur when a key or index used on a mapping or sequence is invalid. It serves as the base class for the KeyError and IndexError exceptions.
Let’s take a look at each of these exceptions with an example.

1 KeyError: This exception is raised when you try to access a key in a dictionary that does not exist.

```python
my_dict = {"apple": 1, "banana": 2}
print(my_dict["orange"])  # Raises KeyError: 'orange' not found in my_dict
```

2 IndexError: This exception occurs when you try to access an index in a sequence (such as a list or tuple) that is out of range. example is below.

```python
my_list = [1, 2, 3]
print(my_list[3])  # Raises IndexError: list index out of range
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans:- ImportError:- occurs when the Python program tries to import a module, its classes, or its methods, but fails to do so. This can happen due to various reasons such as an invalid or incorrect path, a faulty installation of an external library, or a circular dependency.To handle this exception, you can use exception handling with try and except blocks.

ModuleNotFoundError is a subclass of ImportError that is raised when the specified module does not exist.It helps you identify if the module you are trying to import is missing.
For example, if you try to import a module named “request” that is not present in your Python library, you will encounter an ImportError.



Q6. List down some best practices for exception handling in python.

Ans:- Certainly! Here are some best practices for exception handling in Python:

1 Use exceptions for exceptional situations:- Exceptions should be used to handle events that are not part of normal execution. 
For example, when searching for a pattern in a string, returning -1 if the pattern isn’t found is considered normal execution, while indexing beyond the end of a string raises an exception.

2 Handle exceptions at the appropriate level:- Exceptions should be handled at the level that knows how to handle them. Always ask yourself, "Is this the right place to handle this exception?".

3 Avoid using exceptions for flow control:- Exceptions should not be used for flow control. They exist to handle exceptional situations, not as a substitute for if-else statements or loops.

Design for failure: Plan ahead by considering possible failures and designing your program to handle them gracefully. This means anticipating edge cases and implementing appropriate error handlers2
